```python
# -*- coding: utf-8 -*-
"""
Created on Thu May 10 10:45:37 2018

@author: dedekinds
"""

import os
import pandas as pd
import random
import numpy as np
import matplotlib.pyplot as plt
import lightgbm as lgb
from sklearn.externals import joblib
from sklearn.ensemble import BaggingRegressor
from sklearn.model_selection import train_test_split
#LD_data_8_60_time_new.csv

location = 'BJ'
num_of_variable =[145,174][location == 'BJ']


f = open('/home/dedekinds/BJ_train_data_set_2017-01_2018-05-20_without_NO2_without_weather.csv')#BJ_all_month_aqi_meo_75_model_train_new.csv
#BJ_train_data_set_2017-01_2018-05_without_weather.csv
def smape(actual, predicted):
    a = np.abs(np.array(actual) - np.array(predicted))
    b = np.array(actual) + np.array(predicted)
    result = 2 * np.mean(np.divide(a, b, out=np.zeros_like(a), where=b!=0, casting='unsafe'))
    return 'smape',[23333,result][result>=0],0

def num_out_val(location):
    if location == 'LD':return 2
    else:return 3
out_size = 60*num_out_val(location)    
def splitdata(data):
    #按时间顺序来split数据集变为测试集和训练集
    #per:测试集占比    data数据集(array)
    test_data_raw = int(data.shape[0]*p)
    n = data.shape[0]
    m = test_data_raw
    #在n个数中不重复抽取m个样本且保证顺序哦~ 很适合时间序列的说
    delete_row = []
    #记下来data被抽取出来的行的index
    for i in range(n):
        if random.randint(0, n-i+1) < m:
            delete_row.append(i)
            m-=1

    data = np.delete(data, delete_row, axis=0) 
    return data #test_data train_data


df =pd.read_csv(f)
df = df.dropna(axis=0,how='any')
temp_data = df.values
per = 0.03
p = 0.5




for j in range(out_size):
    print(j)
    #loca = (0,1)+tuple(range(51,65))#enhance2hour
    X = temp_data[:,:num_of_variable]
    Y = temp_data[:,num_of_variable+j]
    train_x, test_x, train_y, test_y =train_test_split(X,Y,test_size=0.35)
    
    
    num = 5
    merge_model = []
    
    for i in range(1,num+1):
        gbm = lgb.LGBMRegressor(objective='regression',
                                num_leaves=30+i*2,
                                learning_rate=0.3,
                                n_estimators=500)
        gbm.fit(train_x, train_y,
                eval_set=[(test_x, test_y)],
                eval_metric=smape,
                early_stopping_rounds=8)
        merge_model.append(gbm)
        
    model_dir = "aqi48_75"
    model_name = '_lightgbm_'+location+str(j)+'_include_April.txt'#'_lightgbm_'+location+str(model_index)+'_include_April.txt'
    if not os.path.exists(model_dir):
        os.mkdir(model_dir)
    joblib.dump(merge_model, os.path.join(model_dir, model_name))
```
